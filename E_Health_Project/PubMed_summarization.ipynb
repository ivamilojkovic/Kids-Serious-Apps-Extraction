# PubMed Class 

The class contains the following methods:
* __ __init__ __ (self, output_path, sleep_min)
* __search__ (self, keyword)
* __retreive_abstract__ (self, uid, web_env)

```python
class PubMedPaper:

  def __init__(self, output_path, sleep_min = 1):
    self.sleep_minutes = sleep_min
    self.base_url_esearch =  "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?"
    self.base_url_fetch = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?"
    self.output_file_path = output_path


  def search(self, search_term):
    url = self.base_url_esearch + 'db=pubmed&term=' + search_term + "retmode=json&retmax=10&usehistory=y"
    articles_url = urllib.request.urlopen(url)
    
    text_xml = articles_url.read().decode('utf-8')
    articles_xml = BeautifulSoup(text_xml, 'lxml')
    
    articles = articles_xml.get_text()
    articles = list(articles.split('\n'))
    
    id_articles = articles[1:-2]
    web_environment = articles[0]
    print(web_environment)
    
    for id in id_articles:
      self.retrieve_abstract(id, web_environment)

  def retrieve_abstract(self, uid, webenv):
    url = self.base_url_fetch + "db=pubmed&retmode=xml&id=" + uid + "&webenv=" + webenv

    result = urllib.request.urlopen(url)
    xml = result.read().decode('utf-8')
    xml = BeautifulSoup(xml, 'lxml')
    txt_xml = xml.get_text() 
    print(txt_xml)

    filename = self.output_file_path + "/PubMed_" + uid + ".xml"
    print(filename)

    file_out = open(filename, "w")
    file_out.write(txt_xml)
    file_out.close()

    time.sleep(self.sleep_minutes)
```

```python
pubmed = PubMedPaper(output_path)

term = "children+apps+learning"
pubmed.search(term)
```
